Think of yourself, currently working as a Business analyst in one of the top sports company. The senior management team has asked you to come up with metrics with which they can gauge which team will win the upcoming La Liga cup (Football tournament).

The data set contains information on all the teams so far participated in all the past tournaments. It has data about how many goals each team scored, conceded, how many times they came within first 6 positions, how many seasons they have qualified, their best position in the past etc.

 

You are required to do the following:

Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)
Print all the teams which have started playing between 1930-1980. (5 points)
Print the list of teams which came Top 5 in terms of points (2.5 points)
Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

Create a new column with name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
Eg: Best Position                Points

        1                              25000

        2                              7000      

```python
import numpy as np
import pandas as pd
import sys
```

```python
#Reading the CSV File
df=pd.read_csv("Laliga.csv")
df.head(5)
```

```python
#Making first row as column names
laliga_df=pd.DataFrame(np.array(df[1:]), columns=df.iloc[0], copy=True)
laliga_df.tail(5)
# Re-indexing
#Index=np.arange(0,61)
#laliga_df=laliga_df.reindex(Index)
```

```python
#Replacing '-' with '0'
laliga_df.iloc[laliga_df.iloc[:] == '-']=0
#Print few rows (with dashes) after replacement
laliga_df.tail(5)
```

```python
#Change Type of few columns to allow arithmatic comarisions and other operations
laliga_df[['Points', 'GamesPlayed', 'GamesWon', 'GamesDrawn', 'GamesLost', 'GoalsFor', 'GoalsAgainst', 'BestPosition']]=\
laliga_df[['Points', 'GamesPlayed', 'GamesWon', 'GamesDrawn', 'GamesLost', 'GoalsFor', 'GoalsAgainst', 'BestPosition']].apply(pd.to_numeric)
laliga_df.dtypes
```

```python
#Print all the teams which have started playing between 1930-1980
laliga_df[['StartYear', 'EndYear']]=laliga_df['Debut'].str.split("-", n=1, expand=True)
#laliga_df.fillna(0, inplace=True)
laliga_df[['StartYear', 'EndYear']]=laliga_df[['StartYear', 'EndYear']].apply(pd.to_numeric)
laliga_df.loc[(laliga_df['StartYear']>1930)&(laliga_df['StartYear']<1980),('Team','Debut')]
```

```python
#Print the list of teams which came Top 5 in terms of points
laliga_df.loc[:,('Team','Points')].sort_values(by=['Points'],ascending=False).head(5)
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference
#Goal_diff_count = GoalsFor - GoalsAgainst
def Goal_diff_count(x):
    x['Goal Differences']=np.array(x['GoalsFor'])-np.array(x['GoalsAgainst'])
    return x.loc[:,('Team','Goal Differences')]

print(np.max(Goal_diff_count(laliga_df)))
print(np.min(Goal_diff_count(laliga_df)))
```

```python
#Create a new column with name “Winning Percent” and append it to the data set
# Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
try:
    laliga_df['Winning Percent'] = (laliga_df['GamesWon']/laliga_df['GamesPlayed'])*100
    laliga_df.fillna(0,inplace=True)
except RuntimeWarning:
    print("Value invalid")

laliga_df[['Pos', 'Team', 'GamesPlayed', 'GamesWon', 'Winning Percent']].tail(5)
```

```python
#Print the top 5 teams which has the highest Winning percentage
laliga_df.loc[:,('Team','Winning Percent')].sort_values(by=['Winning Percent'],ascending=False).head(5)
```

```python
#Group teams based on their “Best position”
#print the sum of their points for all positions (5 points) Eg: Best Position Points
laliga_df.groupby("BestPosition")['Points'].sum()
```
